```python
import pandas_datareader.data as web
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import matplotlib.ticker as ticker
from matplotlib.ticker import FormatStrFormatter
```

## **Â¿How to download and visualize data from FRED?**


<font size="3"> The Federal Reserve Bank of St Louis, also known as FRED, has one of the most comprehensive databases on financial time series. In this notebook you can find an example on how to get this data and plot it. FRED has an API that you can use to download the data using only python code. The pandas_datareader library is very useful for that purpose Below you can find an example on how this works.<br><br>

```python
#Dates
start = datetime(1990, 1, 1)
end = datetime(2022, 12, 31)

#Instantaneous forward rate
US_FWD_1D1Y= web.DataReader('THREEFF1', 'fred', start, end)
#Term premia
US_TERM_PREMIA_10Y= web.DataReader('THREEFYTP10', 'fred', start, end)
```

```python
#------------------------CHART -----------------------------------#
font = font_manager.FontProperties(family='Impact',weight='bold', style='normal', size=15)
tick_spacing1 = 400

fig = plt.figure()

ax1 = fig.add_subplot(1,1,1)
ax1.plot(US_FWD_1D1Y.index,US_FWD_1D1Y['THREEFF1'],marker='o',linestyle='solid',linewidth=3,color='lime',alpha=0.99,label='INSTANTANEOUS FORWARD RATE 1Y HENCE',markersize=1)
ax1.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(0.75, 1.0),ncol=10,markerscale=10)
ax1.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing1))
plt.xticks(rotation=90)
plt.yticks(fontname = "Tahoma",fontsize=15)
plt.xticks(fontname = "Tahoma",fontsize=15)
ax1.title.set_fontfamily('arial')
ax1.axis([start, end, None, None])
vals = ax1.get_yticks()
ax1.yaxis.set_major_formatter(FormatStrFormatter('%2.2f%%'))
plt.box(on=None)

ax2 = ax1.twinx()
ax2.plot(US_TERM_PREMIA_10Y.index,US_TERM_PREMIA_10Y['THREEFYTP10'],marker='o',linestyle='solid',linewidth=3,color='black',alpha=0.99,label='TERM PREMIA 10Y',markersize=1)
ax2.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(1.0, 1.0),ncol=10,markerscale=10)
ax2.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing1))
plt.yticks(fontname = "Tahoma",fontsize=15)
plt.xticks(fontname = "Tahoma",fontsize=15)
ax2.axis([start, end, None, None])
vals = ax2.get_yticks()
ax2.yaxis.set_major_formatter(FormatStrFormatter('%2.2f%%'))
plt.box(on=None)

plt.title('FACTORS THAT DETERMINE BOND YIELDS ', size=25, family='Tahoma',weight='bold')
plt.subplots_adjust(bottom=0.1, right=1.8, top=1.0)

plt.show()
```
