```python
from pandas_datareader import wb
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import matplotlib.ticker as ticker
from matplotlib.ticker import FormatStrFormatter
```

## **Â¿How to download and visualize data from the World Bank?**


<font size="3"> The World Bank has one of the most comprehensive database about economic and social variables for countries. In this notebook you can find a couple of examples on how to get this data and plot it. In Data Science this process is know as Exploratory Data Analysis - EDA.<br><br>
    
<font size="3"> The World Bank has an API that you can use to download the data using only python code. The pandas_datareader library is very useful for that purpose Below you can find an example on how this works.<br><br>

```python
#Dates
start = datetime(1961, 1, 1)
end = datetime(2022, 12, 31)

#Countries
countries = ['BR','CO','MX','CL','PE','AR','ES']
```

```python
#Mortality rate infants - per 1.000 live birhts
variable = 'SP.DYN.IMRT.IN'   #You can find the codes here -> https://data.worldbank.org/indicator
variable_description = 'Mortality rate infants - per 1.000 live birhts'  

variable_df = wb.download(indicator = variable, start=start,end=end,country = countries)
variable_df.reset_index(inplace=True)

df_TOTAL = variable_df[['country','year',variable]]
df_TOTAL = df_TOTAL.pivot('country','year',variable)
df_TOTAL.reset_index(inplace=True)
df_TOTAL.set_index('country',inplace=True)
df_TOTAL2 = df_TOTAL.T
df_TOTAL2.set_index(np.arange(1961,2022,1),inplace=True)


font = font_manager.FontProperties(family='Tahoma',weight='normal', style='normal', size=15)
tick_spacing1 = 1

fig = plt.figure()

ax1 = fig.add_subplot(1,1,1)
ax1.plot(df_TOTAL2.index,df_TOTAL2['Argentina'],marker='o', linestyle='solid',linewidth=5,color='black',alpha=0.99,label='ARG',markersize=2)
ax1.plot(df_TOTAL2.index,df_TOTAL2['Brazil'],marker='o', linestyle='solid',linewidth=5,color='lime',alpha=0.99,label='BRA',markersize=2)
ax1.plot(df_TOTAL2.index,df_TOTAL2['Mexico'],marker='o', linestyle='solid',linewidth=5,color='darkorange',alpha=0.99,label='MEX',markersize=2)
ax1.plot(df_TOTAL2.index,df_TOTAL2['Peru'],marker='o', linestyle='solid',linewidth=5,color='lightgrey',alpha=0.99,label='PER',markersize=2)
ax1.plot(df_TOTAL2.index,df_TOTAL2['Chile'],marker='o', linestyle='solid',linewidth=5,color='royalblue',alpha=0.99,label='CHI',markersize=2)
ax1.plot(df_TOTAL2.index,df_TOTAL2['Colombia'],marker='o', linestyle='solid',linewidth=5,color='orangered',alpha=0.99,label='COL',markersize=2)
ax1.plot(df_TOTAL2.index,df_TOTAL2['Spain'],marker='o', linestyle='solid',linewidth=5,color='gold',alpha=0.99,label='ESP',markersize=2)

ax1.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing1))
ax1.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(0.9, 0.95),ncol=7,markerscale=5)
plt.xticks(rotation=90)
plt.yticks(fontname = "Tahoma",fontsize=15)
plt.xticks(fontname = "Tahoma",fontsize=15)
ax1.title.set_fontfamily('arial')
ax1.axis([None, None, None, None])
vals = ax1.get_yticks()
ax1.yaxis.set_major_formatter(FormatStrFormatter('%.2f'))
plt.box(on=None)

plt.title(variable_description, size=25, family='Tahoma',weight='bold')
plt.subplots_adjust(bottom=0.1, right=2.5, top=1.5)

plt.show()
```

```python
#GDP GROWTH
variable = 'NY.GDP.MKTP.KD.ZG'
variable_description = 'GDP Growth'  

variable_df = wb.download(indicator = variable, start=start,end=end,country = countries)
variable_df.reset_index(inplace=True)

df_TOTAL = variable_df[['country','year',variable]]
df_TOTAL = df_TOTAL.pivot('country','year',variable)
df_TOTAL.reset_index(inplace=True)
df_TOTAL.set_index('country',inplace=True)

df_TOTAL2 = df_TOTAL.T
df_TOTAL2.set_index(np.arange(1961,2022,1),inplace=True)

nuevo_df = df_TOTAL2

nuevo_df['Argentina_GDP_Acum'] =(1 + nuevo_df['Argentina']/100).cumprod() - 1
nuevo_df['Brazil_GDP_Acum'] =(1 + nuevo_df['Brazil']/100).cumprod() - 1
nuevo_df['Chile_GDP_Acum'] =(1 + nuevo_df['Chile']/100).cumprod() - 1
nuevo_df['Colombia_GDP_Acum'] =(1 + nuevo_df['Colombia']/100).cumprod() - 1
nuevo_df['Mexico_GDP_Acum'] =(1 + nuevo_df['Mexico']/100).cumprod() - 1
nuevo_df['Peru_GDP_Acum'] =(1 + nuevo_df['Peru']/100).cumprod() - 1
nuevo_df['Spain_GDP_Acum'] =(1 + nuevo_df['Spain']/100).cumprod() - 1

font = font_manager.FontProperties(family='Tahoma',weight='bold', style='normal', size=15)
tick_spacing1 = 1

fig = plt.figure()

ax1 = fig.add_subplot(1,1,1)
ax1.plot(nuevo_df.index,nuevo_df['Argentina_GDP_Acum']*100,marker='o', linestyle='solid',linewidth=5,color='black',alpha=0.99,label='ARG',markersize=2)
ax1.plot(nuevo_df.index,nuevo_df['Brazil_GDP_Acum']*100,marker='o', linestyle='solid',linewidth=5,color='lime',alpha=0.99,label='BRA',markersize=2)
ax1.plot(nuevo_df.index,nuevo_df['Mexico_GDP_Acum']*100,marker='o', linestyle='solid',linewidth=5,color='darkorange',alpha=0.99,label='MEX',markersize=2)
ax1.plot(nuevo_df.index,nuevo_df['Peru_GDP_Acum']*100,marker='o', linestyle='solid',linewidth=5,color='lightgrey',alpha=0.99,label='PER',markersize=2)
ax1.plot(nuevo_df.index,nuevo_df['Chile_GDP_Acum']*100,marker='o', linestyle='solid',linewidth=5,color='royalblue',alpha=0.99,label='CHI',markersize=2)
ax1.plot(nuevo_df.index,nuevo_df['Colombia_GDP_Acum']*100,marker='o', linestyle='solid',linewidth=5,color='orangered',alpha=0.99,label='COL',markersize=2)
ax1.plot(nuevo_df.index,nuevo_df['Spain_GDP_Acum']*100,marker='o', linestyle='solid',linewidth=5,color='gold',alpha=0.99,label='ESP',markersize=2)

ax1.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing1))
ax1.legend(frameon=False,prop=font,loc=1, bbox_to_anchor=(0.9, 1.00),ncol=7,markerscale=5)
plt.xticks(rotation=90)
plt.yticks(fontname = "Tahoma",fontsize=15)
plt.xticks(fontname = "Tahoma",fontsize=15)
ax1.title.set_fontfamily('arial')
ax1.axis([None, None, None, None])
vals = ax1.get_yticks()
ax1.yaxis.set_major_formatter(FormatStrFormatter('%2.2f%%'))
plt.box(on=None)

ax1.text(nuevo_df.index[-1], nuevo_df['Argentina_GDP_Acum'].iloc[-1]*100,'Argentina: '+str(np.round((nuevo_df['Argentina_GDP_Acum'].iloc[-1]**(1/61)-1)*100,2))+'%',color='black',fontsize=10,weight='normal')
ax1.text(nuevo_df.index[-1], nuevo_df['Brazil_GDP_Acum'].iloc[-1]*100,'Brazil: '+str(np.round((nuevo_df['Brazil_GDP_Acum'].iloc[-1]**(1/61)-1)*100,2))+'%',color='black',fontsize=10,weight='normal')
ax1.text(nuevo_df.index[-1], nuevo_df['Mexico_GDP_Acum'].iloc[-1]*100,'MÃ©xico: '+str(np.round((nuevo_df['Mexico_GDP_Acum'].iloc[-1]**(1/61)-1)*100,2))+'%',color='black',fontsize=10,weight='normal')
ax1.text(nuevo_df.index[-1], nuevo_df['Peru_GDP_Acum'].iloc[-1]*100,'PerÃº: '+str(np.round((nuevo_df['Peru_GDP_Acum'].iloc[-1]**(1/61)-1)*100,2))+'%',color='black',fontsize=10,weight='normal')
ax1.text(nuevo_df.index[-1], nuevo_df['Chile_GDP_Acum'].iloc[-1]*100,'Chile: '+str(np.round((nuevo_df['Chile_GDP_Acum'].iloc[-1]**(1/61)-1)*100,2))+'%',color='black',fontsize=10,weight='normal')
ax1.text(nuevo_df.index[-1], nuevo_df['Colombia_GDP_Acum'].iloc[-1]*100,'Colombia: '+str(np.round((nuevo_df['Colombia_GDP_Acum'].iloc[-1]**(1/61)-1)*100,2))+'%',color='black',fontsize=10,weight='normal')
ax1.text(nuevo_df.index[-1], nuevo_df['Spain_GDP_Acum'].iloc[-1]*100,'EspaÃ±a: '+str(np.round((nuevo_df['Spain_GDP_Acum'].iloc[-1]**(1/61)-1)*100,2))+'%',color='black',fontsize=10,weight='normal')

plt.title(variable_description, size=25, family='Tahoma',weight='bold')
plt.subplots_adjust(bottom=0.1, right=2.5, top=1.5)

plt.show()
```
